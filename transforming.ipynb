```python
import pandas as pd
import pytz

plant_df = pd.read_json("sample.json")
plant_df["plant_name"] = plant_df["botanist"].apply(lambda x: x["name"])
plant_df["email"] = plant_df["botanist"].apply(lambda x: x["email"])
plant_df["phone"] = plant_df["botanist"].apply(lambda x: x["phone"])
plant_df = plant_df.drop("botanist", axis=1)
plant_df = plant_df.drop("images", axis=1)
```

```python
gmt = pytz.timezone("Europe/London")

plant_df["recording_taken"] = pd.to_datetime(plant_df["recording_taken"]).dt.tz_localize(pytz.utc)
plant_df["recording_taken"] = plant_df["recording_taken"].dt.tz_convert(gmt)
plant_df["recording_taken"]

plant_df.dtypes
plant_df["recording_taken"]
```

```python
plant_df["last_watered"] = pd.to_datetime(plant_df["last_watered"]).dt.tz_localize(gmt)
plant_df
```

```python
sunlight_types = ['full_sun', 'partial_sun', 'full_shade']
plant_df["sunlight"] = plant_df["sunlight"].apply(lambda x: x[0].lower().replace(" ", "_") if (isinstance(x, list) and x[0].lower().replace(" ", "_") in sunlight_types) else None)
plant_df
#The else NONE could be changed ?
```
